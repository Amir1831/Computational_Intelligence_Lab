Build a Convolutional Network class with two Convolutional layers and one fully connected layer. add Batch Norm for the convolutional layers.

Constructor  using Batch Norm 

```python
class CNN(nn.Module):
    
    # Contructor
    def __init__(self, out_1=16, out_2=32,number_of_classes=10):
        super(CNN, self).__init__()
        # start your code
        self.stack = nn.Sequential(
            #start your code
            nn.Conv2d(1,out_1,3,1,1),
            nn.BatchNorm2d(out_1),
            nn.ReLU(),
            nn.Conv2d(out_1,out_2,3,1,1),
            nn.BatchNorm2d(out_2),
            nn.ReLU(),
            # nn.MaxPool2d(2,2),
            nn.Flatten(),
            nn.Linear(8192 , number_of_classes)
            # nn.ReLU(),
            # nn.Linear(512 , 24),
            # nn.ReLU(),
            # nn.Linear(24 ,2 )
            # end
        )
        
        # end

    
    # Prediction
    def forward(self, x):
        # start your code
        x = self.stack(x)
        
        # end
        return x
```

train loader  and validation loader 

```python
train_loader = DataLoader(dataset_train , batch_size=4 , shuffle = True)
test_loader = DataLoader(dataset_val , batch_size=4 , shuffle = True)
```

Convolutional Neural Network object 

```python
#model = CNN(out_1=16, out_2=32,number_of_classes=10)
model = CNN()
model.to("cuda")
```

Create the objects for the criterion and the optimizer named <code>criterion</code> and <code>optimizer</code>. Make the optimizer use SGD with a learning rate of 0.1 and the optimizer use Cross Entropy Loss

```python
lossfunc = nn.CrossEntropyLoss()
learning_rate = 0.1
optimizer = torch.optim.SGD(model.parameters() , lr = 0.1)
```

Code used to train the model 

```python
len(train_loader) , len(train_loader.dataset)
```

```python
%%time
import time
start_time = time.time()

cost_list=[]
accuracy_list=[]


n_epochs=5

# start your code
for i in range(n_epochs):
    model.train()
    trainloss = 0
    correct = 0

    for img,label in train_loader:
        '''
            Traning the Model.
        '''
        optimizer.zero_grad()     #clear gradient for next train
        img = img.to("cuda")
        label = label.to("cuda")
        output = model(img) # predict

        loss = lossfunc(output , label) #compute loss
        loss.backward()   #backpropagation
        optimizer.step()    #apply gradients

        trainloss+=loss.item()
        
        correct += (output.argmax(1) == label).type(torch.float).sum().item()


    cost_list.append(trainloss/len(train_loader))
    accuracy_list.append(correct/len(train_loader.dataset))




# end
    
    
```
